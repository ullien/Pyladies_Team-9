```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #visualisation
%matplotlib inline 


# read the data and store it in df variable
path = 'data_team-9/covid-19.csv'
covid19_df = pd.read_csv(path)
```

```python
path = 'data_team-9/bcg.csv'
bcg_df = pd.read_csv(path, header=1)
```

```python
path = 'data_team-9/life_expectancy.csv'
life_expectance_df = pd.read_csv(path)
```

```python
covid19_df.columns
```

```python
covid19_df['life expectancy at birth'].describe()
```

```python
#create a dataframe with only necessary columns
mini_covid19_df = covid19_df[['Country_Region','bcg_coverage','bcg_year_delta', 'life expectancy at birth']]
print(mini_covid19_df)
```

```python
#remove the rows (countries) with missing data  
mini_covid19_df_without_na = mini_covid19_df.dropna()
print(mini_covid19_df_without_na)
```

```python
mini_covid19_df_without_na['bcg_year_delta'].describe()
```

```python
# show distribution of the results
plt.scatter(mini_covid19_df_without_na['bcg_coverage'], mini_covid19_df_without_na['life expectancy at birth'], color = 'blue')
plt.title("BCG vaccine vs life expectancy")
plt.xlabel("BCG coverage")
plt.ylabel("Life expectancy at birth")
plt.show()
```

```python
# show that even countries with high bcg coverage have lower life expectancy
mini_covid19_df_without_na.loc[(mini_covid19_df_without_na['life expectancy at birth'] <= 60), ['Country_Region', 'bcg_coverage']]
```

```python
# show that even countries with lower bcg coverage have higher life expectancy
mini_covid19_df_without_na.loc[(mini_covid19_df_without_na['bcg_coverage'] <= 40), ['Country_Region', 'life expectancy at birth']]
```
